```python
#Challenge 2: Summation of Primes
#The sum of the primes below 10 is 2 + 3 + 5 + 7 = 17. 
#Find the sum of all the primes below 2,000. 

sum_1_digit_primes = (2+3+5) #This was the easiest way I could think to include the sum of these prime numbers in my total

prime_check = ["1","3","7","9"] #With the exception of 2 and 5, all prime numbers must end in 1,3,7 or 9
prime_sum = 0
for nums in range(2,2000):
    if str(nums).endswith(tuple(prime_check)) and nums %3 !=0: #The modulus excludes 9 from being counted
        prime_sum = prime_sum + nums
print(prime_sum+sum_1_digit_primes)
    
```

```python
##Challenge 3: Multiples of 3 and 5
##If we list all of the natural numbers below 10 that are multiples of 3 or 5, we get 3, 5, 6, and 9. 
##The sum of these multiples is 23. 
##Find the sum of all the multiples of 3 and 5 below 1,000. 
##Afterward, write a brief explanation walking through your code's logic in markdown.

result = 0
for num in range(1,1000): 
    if num%15 == 0: #All multiples of 3 AND 5 are multiples of 15
        result += num
print(result)
```
